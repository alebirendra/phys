```python
import tkinter as tk
from tkinter import *
import numpy as np
from tkinter import ttk


from PhysicsFunc import Vector

root = Tk()


canvasA = LabelFrame(root, width = 225, height = 625)
canvasA.pack(fill=BOTH, expand=YES, side=LEFT)   
canvasB = LabelFrame(root, width = 625, height = 625, bg='light cyan',  borderwidth=1, highlightthickness=2)
canvasB.pack(fill=BOTH, expand=YES, side=RIGHT, anchor=E) 


dist = ttk.Notebook(canvasB, width=850, height=625)
dist.grid(row=1, column=0, pady=5)

        # create frames
example1 = ttk.Frame(dist) #  , width=1300, height=700
example2 = ttk.Frame(dist)
example3 = ttk.Frame(dist)

example1.pack(fill='both', expand=True)
example2.pack(fill='both', expand=True)
example3.pack(fill='both', expand=True)
dist.add(example1, text='Example 1')
dist.add(example2, text='Example 2')
dist.add(example3, text='Example 3')

lbl_title = Label(canvasA, text='Distance vs Displacement', font=('Arial', 14), fg='blue')
lbl_title.grid(row=0, column=0, columnspan=2, pady=20)

lbl_dist = Label(canvasA, text='Each step along X or Y = 10m', font=('Arial', 14))
lbl_dist.grid(row=1, column=0, columnspan=2, pady=20)
lbl_dist = Label(canvasA, text='Each diagonal step = 14.14m', font=('Arial', 14))
lbl_dist.grid(row=2, column=0, columnspan=2, pady=5)

lbl_dist = Label(canvasA, text='Distance (m)', font=('Arial', 14))
lbl_dist.grid(row=3, column=0, pady=50)
entr_dist = Entry(canvasA, text='', bd=3, width=5, fg='blue', font=('Arial', 14))
entr_dist.grid(row=3, column=1, pady=50)

lbl_disp = Label(canvasA, text='Displacement (m)', font=('Arial', 14))
lbl_disp.grid(row=4, column=0, pady=10)
entr_disp = Entry(canvasA, text='', bd=3, width=5, fg='red',  font=('Arial', 14))
entr_disp.grid(row=4, column=1, pady=10)
lbl_ang = Label(canvasA, text='Angle (degree)', font=('Arial', 14))
lbl_ang.grid(row=5, column=0, pady=5)
entr_ang = Entry(canvasA, text='', bd=3, width=5, fg='red',  font=('Arial', 14))
entr_ang.grid(row=5, column=1, pady=5)

lbl_xcompo = Label(canvasA, text='X Component (m)', font=('Arial', 14))
lbl_xcompo.grid(row=6, column=0, pady=10)
entr_xcompo = Entry(canvasA, text='', bd=3, width=5, fg='red',  font=('Arial', 14))
entr_xcompo.grid(row=6, column=1, pady=10)

lbl_ycompo = Label(canvasA, text='Y Component (m)', font=('Arial', 14))
lbl_ycompo.grid(row=7, column=0, pady=10)
entr_ycompo = Entry(canvasA, text='', bd=3, width=5, fg='red',  font=('Arial', 14))
entr_ycompo.grid(row=7, column=1, pady=10)

def angles( x, y):
    global theta

    if x > 0 and y == 0:
        theta = 0
    if x > 0 and y > 0:
        theta = np.arctan(y/x)
    if x == 0 and y > 0:
        theta = np.pi/2
    if x < 0 and y > 0:
        theta = np.pi/2 + np.arctan(abs(x)/y)
    if x < 0 and y == 0:
        theta = np.pi
    if x < 0 and y < 0:
        theta = np.pi + np.arctan(abs(y)/abs(x))
    if x == 0 and y < 0:
        theta = 3*np.pi/2
    if x > 0 and y < 0:
        theta = 3*np.pi/2 + np.arctan(x/abs(y))                                 
    theta2 = theta*180/np.pi
    return theta2

#...........................ex1..............................................

c_width,c_height = 750, 550 

c3 = Canvas(example3, width=c_width, height=c_height,bg='white')  # ,bg='lightgreen'
c3.grid(row=0,column=0,padx=5,pady=5,columnspan=5)
w_rect = 20
h_rect = 20
sx = 0   # space in x dir
sy = 10   # space in y dir
linewidth = 3

xstep=10  
ystep=10
x1,y1,x2,y2=c_width/2+sx,c_height,w_rect,c_height/2-w_rect  
rx1, ry1, rx2, ry2 = c_width/2+sx-w_rect/2, c_height/2+h_rect/2, c_width/2+sx+w_rect/2, c_height/2-h_rect/2  

lx = c3.create_line(x1-c_width/2,y1/2,c_width,y1/2,arrow='last', width=linewidth) # x Axis 
ly = c3.create_line(x1,y1,x1,5,arrow='last', width=linewidth) # Y  Axis 

to = c3.create_text(x1,y1+10, text= "O",fill="black",font=('Calibri 15 bold'))
tx = c3.create_text(c_width-10,y1/2+10, text= "X",fill="black",font=('Calibri 15 bold'))
ty = c3.create_text(x1+10,10, text= "Y",fill="black",font=('Calibri 15 bold'))

def1 = c3.create_text(c_width/2, c_height/2+13*ystep, text= "Distance can never be negative.",fill="blue",font=('Calibri 15'))
def2 = c3.create_text(c_width/2,c_height/2+ 21*ystep, text= "Displacement is a vector, it can be zero or negative. \n Negative in a sence of direction with a reference point. \n However magnitude of a vector is always zero or positive. \n We can always decompose a vector into X and Y components. \n Look at the display boxes and see how the sign changes.",fill="red",font=('Calibri 15'))

ax1, ay1, ax2,ay2= c_width/2+sx, c_height/2, c_width/2+sx, c_height/2  # initial position of line
bx1, by1, bx2,by2= c_width/2+sx, c_height/2, c_width/2+sx, c_height/2 

r1=c3.create_rectangle(rx1, ry1, rx2, ry2,fill='cyan')  
la=c3.create_line(ax1, ay1, ax2,ay2, fill='blue')
lb=c3.create_line(bx1, by1, bx2,by2, fill='red')

d3 = 0 # initial distance
dp3 =0 #initial disp
ang3=0
xc3=0
yc3=0
def right3():    
    global rx1,ry1,rx2,ry2,r1, ax1, ay1, ax2,ay2,bx1, by1, bx2,by2, la, lb, d3, dp3, ang3
    
    td = c3.create_text(c_width/2,3*ystep, text= "Blue line represents 'Distance'",fill="blue",font=('Calibri 15'))
    tdp = c3.create_text(c_width/2,5*ystep, text= "Red line represents 'Displacement'",fill="red",font=('Calibri 15'))

    if ( rx2<(c_width-xstep)):   
        rx1=rx1+xstep 
        rx2=rx2+xstep 
        ax1=ax2
        ax2=ax2+xstep
        ay1=ay2
        ay2=ay2
        bx2=ax2
        by2=ay2

        c3.delete(r1)  
        c3.delete(lb)
        la=c3.create_line(ax1, ay1, ax2,ay2,fill='blue',width=linewidth)
        lb=c3.create_line(c_width/2+sx, c_height/2, bx2,by2,fill='red', width=linewidth)
        r1=c3.create_rectangle(rx1, ry1,rx2,ry2,fill='cyan')
        d3 = d3 + float(xstep)
        entr_dist.delete(0, 'end')
        entr_dist.insert(END, d3)
        
        dp3 = np.sqrt((bx2-c_width/2-sx)**2+(by2-c_height/2)**2)
        entr_disp.delete(0, 'end')
        entr_disp.insert(END, dp3)

        #myvec = Vector(x=(bx2-sx), y=-(by2-c_height/2), theta=None, magnitude=None)
        ang3 = angles( x=(bx2-c_width/2-sx), y=-(by2-c_height/2))        
        entr_ang.delete(0, 'end')
        entr_ang.insert(END, ang3)
        entr_xcompo.delete(0, 'end')
        entr_ycompo.delete(0, 'end')
        entr_xcompo.insert(END, (bx2-c_width/2-sx))
        entr_ycompo.insert(END, -(by2-c_height/2))
        
        return d3
def left3():    

    global rx1,ry1,rx2,ry2,r1, ax1, ay1, ax2,ay2,bx1, by1, bx2,by2, la, lb, d3, dp3, ang3
    if(rx1 > xstep):  
        rx1=rx1-xstep 
        rx2=rx2-xstep
        
        ax1=ax2
        ax2=ax2-xstep
        ay1=ay2
        ay2=ay2
        bx2=ax2
        by2=ay2
        c3.delete(r1) 
        c3.delete(lb)
        la=c3.create_line(ax1, ay1, ax2,ay2,fill='blue',width=linewidth)
        lb=c3.create_line(c_width/2+sx, c_height/2, bx2,by2,fill='red', width=linewidth)
        r1=c3.create_rectangle(rx1, ry1,rx2,ry2,fill='cyan')
        d3 = d3 + float(xstep)
        entr_dist.delete(0, 'end')
        entr_dist.insert(END, d3)
        
        dp3 = np.sqrt((bx2-c_width/2-sx)**2+(by2-c_height/2)**2)
        entr_disp.delete(0, 'end')
        entr_disp.insert(END, dp3)

        #myvec = Vector(x=(bx2-sx), y=-(by2-c_height/2), theta=None, magnitude=None)
        ang3 = angles( x=(bx2-c_width/2-sx), y=-(by2-c_height/2))        
        entr_ang.delete(0, 'end')
        entr_ang.insert(END, ang3)
        entr_xcompo.delete(0, 'end')
        entr_ycompo.delete(0, 'end')
        entr_xcompo.insert(END, (bx2-c_width/2-sx))
        entr_ycompo.insert(END, -(by2-c_height/2))
        return d3
        
def up3():    
    

    global rx1,ry1,rx2,ry2,r1, ax1, ay1, ax2,ay2,bx1, by1, bx2,by2, la, lb, d3, dp3, ang3
    if(ry1 > ystep):  
        ry1=ry1-ystep 
        ry2=ry2-ystep 
        ay1=ay2
        ay2=ay2-ystep
        ax1=ax2
        ax2=ax2
        bx2=ax2
        by2=ay2
        c3.delete(r1)  
        c3.delete(lb)
        la=c3.create_line(ax1, ay1, ax2,ay2,fill='blue',width=linewidth)
        lb=c3.create_line(c_width/2+sx, c_height/2, bx2,by2,fill='red', width=linewidth)
        r1=c3.create_rectangle(rx1, ry1,rx2,ry2,fill='cyan')
        d3 = d3 + float(xstep)
        entr_dist.delete(0, 'end')
        entr_dist.insert(END, d3)
        
        dp3 = np.sqrt((bx2-c_width/2-sx)**2+(by2-c_height/2)**2)
        entr_disp.delete(0, 'end')
        entr_disp.insert(END, dp3)

        #myvec = Vector(x=(bx2-sx), y=-(by2-c_height/2), theta=None, magnitude=None)
        ang3 = angles( x=(bx2-c_width/2-sx), y=-(by2-c_height/2))        
        entr_ang.delete(0, 'end')
        entr_ang.insert(END, ang3)
        entr_xcompo.delete(0, 'end')
        entr_ycompo.delete(0, 'end')
        entr_xcompo.insert(END, (bx2-c_width/2-sx))
        entr_ycompo.insert(END, -(by2-c_height/2))
        return d3
def down3():    


    global rx1,ry1,rx2,ry2,r1, ax1, ay1, ax2,ay2,bx1, by1, bx2,by2, la, lb, d3, dp3, ang3
    if(ry2 < c_height-ystep):  
        ry1=ry1+ystep 
        ry2=ry2+ystep
        ay1=ay2
        ay2=ay2+ystep
        ax1=ax2
        ax2=ax2
        bx2=ax2
        by2=ay2
        c3.delete(r1) 
        c3.delete(lb)
        la=c3.create_line(ax1, ay1, ax2,ay2,fill='blue',width=linewidth)
        lb=c3.create_line(c_width/2+sx, c_height/2, bx2,by2,fill='red', width=linewidth)
        r1=c3.create_rectangle(rx1, ry1,rx2,ry2,fill='cyan')
        d3 = d3 + float(xstep)
        entr_dist.delete(0, 'end')
        entr_dist.insert(END, d3)
        
        dp3 = np.sqrt((bx2-c_width/2-sx)**2+(by2-c_height/2)**2)
        entr_disp.delete(0, 'end')
        entr_disp.insert(END, dp3)

        #myvec = Vector(x=(bx2-sx), y=-(by2-c_height/2), theta=None, magnitude=None)
        ang3 = angles( x=(bx2-c_width/2-sx), y=-(by2-c_height/2))        
        entr_ang.delete(0, 'end')
        entr_ang.insert(END, ang3)
        entr_xcompo.delete(0, 'end')
        entr_ycompo.delete(0, 'end')
        entr_xcompo.insert(END, (bx2-c_width/2-sx))
        entr_ycompo.insert(END, -(by2-c_height/2))
        return d3
        
b1=tk.Button(example3,text='Left',command=left3)
b1.bind(example3, left3)
b1.grid(row=1,column=1,rowspan=2,padx=0, sticky=E)
b2=tk.Button(example3,text='Up',command=up3)
b2.bind(example3,up3)
b2.grid(row=1,column=2)
b3=tk.Button(example3,text='Right',command=right3)
b3.bind(example3, right3)
b3.grid(row=1,column=3,rowspan=2, padx =0, sticky='W')
b4=tk.Button(example3,text='Down',command=down3)
b4.bind(example3, down3)
b4.grid(row=2,column=2)



def clear():
    global rx1,ry1,rx2,ry2,r1, ax1, ay1, ax2,ay2,bx1, by1, bx2,by2, la, lb, d3, dp3, ang3
    entr_dist.delete(0, 'end')
    entr_disp.delete(0, 'end')
    entr_ang.delete(0, 'end')
    entr_xcompo.delete(0, 'end')
    entr_ycompo.delete(0, 'end')

    c3.delete("all")
    d3=0
    x1,y1,x2,y2=c_width/2+sx,c_height,w_rect,c_height/2-w_rect  
    rx1, ry1, rx2, ry2 = c_width/2+sx-w_rect/2, c_height/2+h_rect/2, c_width/2+sx+w_rect/2, c_height/2-h_rect/2  

    lx = c3.create_line(x1-c_width/2,y1/2,c_width,y1/2,arrow='last', width=linewidth) # x Axis 
    ly = c3.create_line(x1,y1,x1,5,arrow='last', width=linewidth) # Y  Axis 

    to = c3.create_text(x1,y1+10, text= "O",fill="black",font=('Calibri 15 bold'))
    tx = c3.create_text(c_width-10,y1/2+10, text= "X",fill="black",font=('Calibri 15 bold'))
    ty = c3.create_text(x1+10,10, text= "Y",fill="black",font=('Calibri 15 bold'))

    def1 = c3.create_text(c_width/2, c_height/2+13*ystep, text= "Distance can never be negative.",fill="blue",font=('Calibri 15'))
    def2 = c3.create_text(c_width/2,c_height/2+ 21*ystep, text= "Displacement is a vector, it can be zero or negative. \n Negative in a sence of direction with a reference point. \n However magnitude of a vector is always zero or positive. \n We can always decompose a vector into X and Y components. \n Look at the display boxes and see how the sign changes.",fill="red",font=('Calibri 15'))

    ax1, ay1, ax2,ay2= c_width/2+sx, c_height/2, c_width/2+sx, c_height/2  # initial position of line
    bx1, by1, bx2,by2= c_width/2+sx, c_height/2, c_width/2+sx, c_height/2 

    r1=c3.create_rectangle(rx1, ry1, rx2, ry2,fill='cyan')  
    la=c3.create_line(ax1, ay1, ax2,ay2, fill='blue')
    lb=c3.create_line(bx1, by1, bx2,by2, fill='red')






clear_btn = Button(canvasA, text='Clear all', command=clear)
clear_btn.bind(canvasA, clear)
clear_btn.grid(row=8, column=1, padx=10, pady=10)












root.mainloop()
```
